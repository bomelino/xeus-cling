```c++
#pragma cling add_include_path("../include")
#include <iostream>
#include "algoviz/SVG.hpp"
#include <vector>

using namespace std;

AlgoViz::clear();



SVG zeichnung = SVG(500,500, "Teilaufgabe 3");

int rad = 5; 
int speed = 15; 

// benutzer kreis
Circle benutzer = Circle(250,250, rad, &zeichnung); 
benutzer.setFill("red");

// verfolger kreis
Circle verfolger = Circle(200,250, rad, &zeichnung);
verfolger.setFill("blue");

int dx = speed;
int dy = 0;

// koordinaten in zwei vektoren speichern
vector<int> x;
vector<int> y;
```

```c++
int distance(Circle &a, Circle&b){
    int dx = (a.getX()-b.getX());
    int dy = (a.getY()-b.getY());
    return sqrt(dx*dx+dy*dy);
}
```

```c++
std::string taste = "";
int neues_x;
int neues_y;
// hauptprogramm
do{
    // nicht zu schnell zeichnen
    AlgoViz::sleep(60);
    
    
    taste = zeichnung.lastKey();
    // welche richtung bewegt sich der benutzer?
    if (taste == "ArrowLeft") {
        dx = -speed;
        dy = 0;
    } else if (taste == "ArrowRight") {
        dx = speed;
        dy = 0;
    } else if ( taste == "ArrowUp") {
        dx = 0;
        dy = -speed;
    } else if (taste == "ArrowDown") {
        dx = 0;
        dy = speed;
    } else 
    
    // benutzer bewegen 
    neues_x = (benutzer.getX() + dx + 500)%500;
    neues_y = (benutzer.getY() + dy + 500)%500;
    x.push_back(benutzer.getX());
    y.push_back(benutzer.getY());
    
    benutzer.moveTo(neues_x,neues_y);   
    
    if (distance(benutzer,verfolger) > 50.0){
        // verfolger auf die n√§chste position bewegen
        verfolger.moveTo(x[0], y[0]);
        x.erase(x.begin());
        y.erase(y.begin());        
    }
    
} while (taste != "x")
```

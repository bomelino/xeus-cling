Im folgenden wollen wir ein Array verwenden, um die Grundlage für einen Spiele-Klassiker zu legen: Snake!

Wir beginnen damit einen roten Kreis als Kopf mit der Tastatur zu steuern. Im Grunde handelt es sich dabei nur um eine gelenkte Version des fliegenden Kreises. Die Richtung in die sich der Kopf bewegt geben wir durch zwei Variablen `dx` und `dy` vor, die die Bewegungsweite in x- und y-Richtung beschreiben. In jeder Runde werden die beiden Werte dann auf die aktuelle Position aufaddiert. D,h, wenn `dx = 10`und `dy= 0` ist bewegt sich der Kopf in jeder Runde um 10 Einheiten nach rechts.

Um den neuen Ort zu berechnen, speichern wir diesmal seine Koordinaten allerdings nicht selber. Stattdessen "fragen" wir den Kreis selbst nach seinem Ort. Dazu können wir die Befehle `getX()`und
`getY()` verwenden, die die jeweiligen Koordinaten zurückgeben.

```c++
#include <algoviz/SVG.hpp>

using namespace std;

AlgoViz::clear();

string taste = "";

int wait = 100;  // Die Zeit zwischen zwei Schritten
int dx = 10;     // Die anfängliche Richtung
int dy = 0;

SVG spielfeld = SVG(400,400);

// Der Kopf ist rot
Circle kopf = Circle (200,200,10,&spielfeld);
kopf.setFill("red");

// Die Hauptschleife
while ( taste != "x" ) {
    taste = spielfeld.lastKey();
    
    // Frage die Tasten ab
    if ( taste == "ArrowUp") {
        dx = 0;
        dy = -10;
    } // ... Hier fehlen die anderen Tasten
    
    // Bewege den Kopf
    // Dazu verwende die aktuellen Koordinaten und verändere sie.
    kopf.moveTo( kopf.getX()+dx , kopf.getY()+dy );
    
    AlgoViz::sleep(wait);
}
```

Soweit entspricht das Ganze im Wesentlichen dem, was wir schon gemacht haben. Damit jetzt Snake daraus wird, benötigen wir den Körper unserer Schlange oder Raupe. Den möchten wir aus einer Reihe von weiteren grünen Kreisen darstellen.

![Die Raupe](/user-redirect/algoviz/img/snake_knick.png)

In unserem Beispiel soll die Raupe insgesamt zehn Segmente lang sein. Ein Kopf und neun Körper. Um diese zu verwalten bietet sich ein Array von Objecte nder Klasse `Circle` an. Legen wir diese mal an, um zu sehen, wie das geht.

```c++
#include <algoviz/SVG.hpp>
using namespace std;

AlgoViz::clear();

SVG spielfeld = SVG(400,400);

// Das Array mit zehn Kreisen
Circle raupe[10];
```

Wenn wir diesen Abschnitt auführen, sieht man leider nichts. Die Kreise wurden zwar schon mit der Deklaration des Arrays anglegt. Aber sie haben noch keine Größe oder Farbe, und noch viel wichtiger, sie sind noch nicht zum `spielfeld` hinzugefügt worden.

Das alles holen wir in einer For-Schleife nach.

```c++
for ( int segment = 0; segment < 10 ; segment = segment + 1 ) {
    
    // Füge das Segment dem Spielfeld hinzu
    raupe[segment].addTo(&spielfeld);
    
    // Setze die Größe und Farbe
    raupe[segment].setRadius(10);
    raupe[segment].setFill("darkgreen");
    
    // Setze den Ort
    raupe[segment].moveTo(200 + segment * 10, 200);    
}
```

Jetzt sind alle Segmente zu sehen. Allerdings ist der Kopf noch nicht rot. Also färben wir ihn rot. Aber welches Segement sollte der Kopf sein? Die "natürliche" Wahl wäre der Kreis mit Index 0.
Probieren wir es aus.

```c++
raupe[0].setFill("red");
```

Hmm. Das ist nicht ganz optimal. Also nehmen wir mal das letzte Segment.

```c++
raupe[0].setFill("darkgreen"); // Wieder zurück
raupe[9].setFill("red");
```

Das sieht besser aus. Also nehmen wir als Kopf den Kreis an Index 9. Bauen wir das Ganze jetzt mal mit der Tastatursteuerung für den Kopf zusammen. Dabei haben wir keine eigenen Variable für den Kopf, sondern merken uns in der `int`-Variable `kopf`nur den Index des Kopfes unter den Segmenten (also `kopf = 9`.)

```c++
#include <algoviz/SVG.hpp>
using namespace std;

AlgoViz::clear();

// Alles zur Steuerung
string taste = "";

int wait = 100;  // Die Zeit zwischen zwei Schritten
int dx = 10;     // Die anfängliche Richtung
int dy = 0;


SVG spielfeld = SVG(400,400);

// Das Array mit zehn Kreisen
Circle raupe[10];

// Der Index de Kopfes
int kopf = 9;

for ( int segment = 0; segment < 10 ; segment = segment + 1 ) {
    
    // Füge das Segment dem Spielfeld hinzu
    raupe[segment].addTo(&spielfeld);
    
    // Setze die Größe und Farbe
    raupe[segment].setRadius(10);
    raupe[segment].setFill("darkgreen");
    
    // Setze den Ort
    raupe[segment].moveTo(200 + segment * 10, 200);    
}

// Der Kopf ist rot
raupe[kopf].setFill("red");


// Die Hauptschleife
// Wir tauschen den Kreis kopf durch raupe[kopf] aus!
while ( taste != "x" ) {
    taste = spielfeld.lastKey();
    
    // Frage die Tasten ab
    if ( taste == "ArrowUp") {
        dx = 0;
        dy = -10;
    } else if ( taste == "ArrowDown") {
        dx = 0;
        dy = 10;
    } else if ( taste == "ArrowLeft") {
        dx = -10;
        dy = 0;
    } else if ( taste == "ArrowRight") {
        dx = 10;
        dy = 0;
    }
    
    // Bewege den Kopf. Wir wählen ihn über den Index.
    // Dazu verwende die aktuellen Koordinaten und verändere sie.
    raupe[kopf].moveTo( raupe[kopf].getX()+dx , raupe[kopf].getY()+dy );
    
    AlgoViz::sleep(wait);
}
```

Das sieht soweit gar nicht schlecht aus. Nur der Körper folgt einfach nicht dem Kopf. Dafür gibt es mindestens zwei Lösungen, die wir im Folgenden vorstellen werden.

## Variante 1: Alle rücken auf!

Die erste Idee besteht einfach darin, dass **alle** Segmente eine Position vorrücken. D.h. das Segment `i` rückt auf die Position des Segmentes `i+1` auf (zur Erinnerung der Schwanz ist 0 und der Kopf 9). Dabei kann das hintere Element das vor ihm liegende nach seiner Position fragen. Das Wichtige dabei sit, dass wir hinten anfangen, damit die Position nicht "verloren" ist.

Also müssen wir das Folgende für alle Segmente außer dem Kopf durchführen:

`raupe[i].moveTo( raupe[i+1].getX(), raupe[i+1].getY() );`

und `raupe[i]` muss **vor** `raupe[i+1]` bewegt werden.

Das kann mit einer einfachen For-Schleife realisiert werden.

```c++
/**
 ** Snake!
 **
 ** Variante 1: alle rücken auf!
 **
 ** Es ändern sich nur ein paar Zeilen.
 **/

#include <algoviz/SVG.hpp>
using namespace std;

AlgoViz::clear();

// Alles zur Steuerung
string taste = "";

int wait = 100;  // Die Zeit zwischen zwei Schritten
int dx = 10;     // Die anfängliche Richtung
int dy = 0;


SVG spielfeld = SVG(400,400);

// Das Array mit zehn Kreisen
Circle raupe[10];

// Der Index des Kopfes
int kopf = 9;

for ( int segment = 0; segment < 10 ; segment = segment + 1 ) {
    
    // Füge das Segment dem Spielfeld hinzu
    raupe[segment].addTo(&spielfeld);
    
    // Setze die Größe und Farbe
    raupe[segment].setRadius(10);
    raupe[segment].setFill("darkgreen");
    
    // Setze den Ort
    raupe[segment].moveTo(200 + segment * 10, 200);    
}

// Der Kopf ist rot
raupe[kopf].setFill("red");


// Die Hauptschleife
// Wir tauschen den Kreis kopf durch raupe[kopf] aus!
while ( taste != "x" ) {
    taste = spielfeld.lastKey();
    
    // Frage die Tasten ab
    if ( taste == "ArrowUp") {
        dx = 0;
        dy = -10;
    } else if ( taste == "ArrowDown") {
        dx = 0;
        dy = 10;
    } else if ( taste == "ArrowLeft") {
        dx = -10;
        dy = 0;
    } else if ( taste == "ArrowRight") {
        dx = 10;
        dy = 0;
    }
    
    // AB HIER NEU
    
    // Die Körpersegmente rücken auf
    for ( int segment = 0; segment < 9 ; segment = segment+1 ) {
        raupe[segment].moveTo( raupe[segment+1].getX(), raupe[segment+1].getY() );
    }
    
    // BIS HIER NEU
    
    // Bewege den Kopf. Wir wählen ihn über den Index.
    // Dazu verwende die aktuellen Koordinaten und verändere sie.
    raupe[kopf].moveTo( raupe[kopf].getX()+dx , raupe[kopf].getY()+dy );
    
    AlgoViz::sleep(wait);
}
```

## Variant 2: Mache den Schwanz zum Kopf

Die erste Variante fnktioniert schon gut. Jedoch werden unnötigerweise alle Segmente verschoben. Dabei müsste man nur eines verschieben, wie die folgende Überlegung zeigt.

Bei der Bewegung muss der Schwanz lediglich an die neue Position des Kopfes verschoben werden. Anschließend muss man noch die Farben anpassen, damit der alte Schwanz zun neuen Kopf wird und der alte Kopf zu einem Körpersegment.

![Der Schwanz wird zum Kopf](/user-redirect/algoviz/img/snake_schritte.png)

Als Quelltext entspricht das in etwa dem Folgenden:

```c++
// ACHTUNG! NICHT ausführbar 

raupe[schwanz].moveTo( raupe[kopf].getX()+dx , raupe[kopf].getY()+dy );
raupe[kopf].setFill("darkgreen");
raupe[schwanz].setFill("red");
```

Anschließend müssten wir `kopf` und `schwanz` anpassen, so dass wir in der nächsten Runde wieder die richtigen Werte haben.

Der neue Kopf ist offensichtlich der alte Schwanz, also `kopf = schwanz`. Und der neue Schwanz ist das Segment vor dem alten, also `schwanz = schwanz + 1`. Aber hier ergibt sich ein Problem. Der Index des Schwanzes wird um 1 erhöht. Aber was passiert, wenn er das Ende des Arrays erreicht? Daz genügt es, wenn wir die ersten drei Schritte ansehen.

![Der Ringpuffer](/user-redirect/algoviz/img/snake_ringpuffer.png)

Die erste Zeile beschriebt die Ausgangssituation. `kopf` ist 9 und der Schwanz bei 0. Wenn wir jetzt einen Schritt machen wird der alte Schwanz zum Kopf, also `kopf = 0` Der Schwanz rückt einen weiter. Im dritten Schritt wird dann `kopf = 1` gesetzt.

Also, wenn `kopf` über das Ende des Arrays hinausläuft, kommt es "vorne" wieder an. D.h. `kopf` läuft 
[im Kreis](/user-redirect/algoviz/lessons/02_Grundlagen/05_RechnenImKreis.ipynb). Das kennen wir schon. Um den nächsten Schritt zu machen machen müssen wir lediglich

`kopf = (kopf + 1) % 10`.

setzen. Den Index des Schwanzes können wir genauso berechnen: `schwanz = (kopf+1) % 10`.

Fassen wir das Ganze im Quelltext zusammen, ergibt sich das folgende Programm.


```c++
/**
 ** Snake!
 **
 ** Variante 2: Der Schwanz wird zum Kopf
 **
 ** Es ändern sich nur ein paar Zeilen.
 **/

#include <algoviz/SVG.hpp>
using namespace std;

AlgoViz::clear();

// Alles zur Steuerung
string taste = "";

int wait = 100;  // Die Zeit zwischen zwei Schritten
int dx = 10;     // Die anfängliche Richtung
int dy = 0;


SVG spielfeld = SVG(400,400);

// Das Array mit zehn Kreisen
Circle raupe[10];

// Der Index des Kopfes
int kopf = 9;

// DAS HIER IST NEU!!
// Eine neue Variable!
int schwanz = 0;

for ( int segment = 0; segment < 10 ; segment = segment + 1 ) {
    
    // Füge das Segment dem Spielfeld hinzu
    raupe[segment].addTo(&spielfeld);
    
    // Setze die Größe und Farbe
    raupe[segment].setRadius(10);
    raupe[segment].setFill("darkgreen");
    
    // Setze den Ort
    raupe[segment].moveTo(200 + segment * 10, 200);    
}

// Der Kopf ist rot
raupe[kopf].setFill("red");


// Die Hauptschleife
// Wir tauschen den Kreis kopf durch raupe[kopf] aus!
while ( taste != "x" ) {
    taste = spielfeld.lastKey();
    
    // Frage die Tasten ab
    if ( taste == "ArrowUp") {
        dx = 0;
        dy = -10;
    } else if ( taste == "ArrowDown") {
        dx = 0;
        dy = 10;
    } else if ( taste == "ArrowLeft") {
        dx = -10;
        dy = 0;
    } else if ( taste == "ArrowRight") {
        dx = 10;
        dy = 0;
    }
    
    // AB HIER NEU
    
    // Der Schwanz wird zum neune Kopf
    raupe[schwanz].moveTo( raupe[kopf].getX() + dx, raupe[kopf].getY() + dy );    
    // Und die farben werden angepasst.
    raupe[schwanz].setFill("red");
    raupe[kopf].setFill("darkgreen");
    
    // Die Indices weren angepasst.
    kopf = schwanz;
    schwanz = (kopf + 1) % 10;
    // BIS HIER NEU
    
    AlgoViz::sleep(wait);
}
```

## Eine mögliche Verbesserung

Die folgende Variante birgt eine Verbesserung gegenüber Variante 1. Dabei nutzen wir das "Rechnen im Kreis", das auch beim Rechnen in beschränkten Zahlräumen verwendet wird. Wir nutzen es, um die Raupe, wenn eine Ihrer Korrdinaten größer als 400 ist und sie aus dem Spielbereich verschwinden würde, am anderen Rand wieder erscheinen zu lassen. D.h. wir nterpretieren Ihre Koordinaten als Kreis mit Werten von 0 bis 399.

Also ergeben sich die neuen Koordinaten des Kopfes als

`raupe[kopf].moveTo( (raupe[kopf].getX()+dx) % 400 , (raupe[kopf].getY()+dy) % 400 );` 

Durch das `% 400` werden zu große Koordinaten wieder zu den entsprechend kleinen Koordinaten umgerechnet.

Allerdings funktioniert das nicht ganz so einfach, wenn die neue Koordinate negativ ist. Daher müssen wir dafür sorgen dass sie vor dem Modulo positiv wird. Das können wir einfach dadurch erreichen, dass wir 400 addieren. Das verschiebt eine negative Koordinate zwischen -4400 und -1 in den positiven Bereich. Durch das Modul wird die 400 aber wieder "entfernt".

`raupe[kopf].moveTo( (raupe[kopf].getX() + dx + 400) % 400 , (raupe[kopf].getY() + dy + 400) % 400 );` 

Ersetzen wir die entsprechende Zeile in Variante 1 durch diese Zeile, erhalten wir eine Variant, bei der die Raupe auf der anderen Seite auftaucht, wenn sie das Spielfeld verlässt. Man sagt auch, dass das Spielfeld ein [**Torus**](https://de.wikipedia.org/wiki/Torus) ist und die Raupe sich auf dem Torus bewegt.

![Torus aus Rechteck (Bildquelle: wikipeduia.de)](/user-redirect/algoviz/img/Torus_from_rectangle.gif)

```c++
/**
 ** Snake!
 **
 ** Variante 3: Die Raupe auf dem Torus
 **
 ** Es ändert sich nur eine Zeile gegenüber Variante 1.
 **/

#include <algoviz/SVG.hpp>
using namespace std;

AlgoViz::clear();

// Alles zur Steuerung
string taste = "";

int wait = 100;  // Die Zeit zwischen zwei Schritten
int dx = 10;     // Die anfängliche Richtung
int dy = 0;


SVG spielfeld = SVG(400,400);

// Das Array mit zehn Kreisen
Circle raupe[10];

// Der Index des Kopfes
int kopf = 9;

for ( int segment = 0; segment < 10 ; segment = segment + 1 ) {
    
    // Füge das Segment dem Spielfeld hinzu
    raupe[segment].addTo(&spielfeld);
    
    // Setze die Größe und Farbe
    raupe[segment].setRadius(10);
    raupe[segment].setFill("darkgreen");
    
    // Setze den Ort
    raupe[segment].moveTo(200 + segment * 10, 200);    
}

// Der Kopf ist rot
raupe[kopf].setFill("red");


// Die Hauptschleife
// Wir tauschen den Kreis kopf durch raupe[kopf] aus!
while ( taste != "x" ) {
    taste = spielfeld.lastKey();
    
    // Frage die Tasten ab
    if ( taste == "ArrowUp") {
        dx = 0;
        dy = -10;
    } else if ( taste == "ArrowDown") {
        dx = 0;
        dy = 10;
    } else if ( taste == "ArrowLeft") {
        dx = -10;
        dy = 0;
    } else if ( taste == "ArrowRight") {
        dx = 10;
        dy = 0;
    }
    
    // Die Körpersegmente rücken auf
    for ( int segment = 0; segment < 9 ; segment = segment+1 ) {
        raupe[segment].moveTo( raupe[segment+1].getX(), raupe[segment+1].getY() );
    }
    
    // Bewege den Kopf. Wir wählen ihn über den Index.
    // Dazu verwende die aktuellen Koordinaten und verändere sie.
    // Die Koordinaten werden dabei "im Kreis gerechnet".
    raupe[kopf].moveTo( ( raupe[kopf].getX() + dx + 400)%400 , ( raupe[kopf].getY() + dy + 400)%400 );
    
    AlgoViz::sleep(wait);
}
```
